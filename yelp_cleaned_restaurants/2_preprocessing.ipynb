# Data Preprocessing

Data preprocessing is an important stage in the data mining pipeline. This notebook focuses on identifying missing values, duplicates, and preparing the dataset for further analysis.

```python
# Import required libraries

import pandas as pd
import numpy as np
```

```python
## Load Dataset
df = pd.read_csv("cleaned_yelp_restaurants.csv")
```

## Missing Value Analysis

```python
# Check missing values

print("Missing Values")
display(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
#Dropping duplicates
df = df.drop_duplicates()

df = df.dropna(subset=['categories'])

print("Shape after cleaning:")
print(df.shape)
```

```python
# Save cleaned dataset for later notebooks
df.to_csv("processed_yelp_restaurants.csv", index=False)

print("Processed dataset saved successfully.")
```
