# Restaurant Recommendation System

This notebook builds a content-based restaurant recommendation system using restaurant categories, TF-IDF vectorization, and cosine similarity.

```python
# Import required libraries
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
## Load Segmented Dataset

df = pd.read_csv("segmented_yelp_restaurants.csv")

df.head()
```

## Feature Preparation

```python
# Creating clean category text field
df['categories_clean'] = df['categories'].fillna('')
```

## TF-IDF Vectorization

Restaurant categories are converted into numerical TF-IDF vectors so that similarity can be calculated.

```python
# Converting restaurant categories into TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['categories_clean'])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

## Recommendation Function

```python
# Create recommendation function using cosine similarity
def recommend_restaurants(restaurant_name, top_n=5):

    # Find restaurants matching the search name
    matches = df[df['name'].str.contains(
        restaurant_name,
        case=False,
        na=False
    )]

    # Return message if restaurant is not found
    if len(matches) == 0:
        return "Restaurant not found"

    # Use first matching restaurant
    idx = matches.index[0]

    # Calculate similarity between selected restaurant and all restaurants
    similarity_scores = cosine_sim[idx]

    # Sort restaurants by similarity score
    similar_indices = similarity_scores.argsort()[::-1]

    # Remove selected restaurant itself and return top recommendations
    similar_indices = [
        i for i in similar_indices
        if i != idx
    ][:top_n]

    return df.iloc[similar_indices][
        ['name', 'city', 'state', 'stars', 'review_count', 'categories']
    ]
```

```python
# Test recommendation system
recommend_restaurants("Sonic", top_n=5)
```

## Cosine Similarity Matrix

Cosine similarity is used to measure the similarity between restaurants based on their category vectors. Restaurants with similar categories will have similarity scores closer to 1, while unrelated restaurants will have scores closer to 0.

```python
# Import cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Create cosine similarity matrix

cosine_sim = cosine_similarity(tfidf_matrix)

print("Cosine Similarity Matrix Shape:")
print(cosine_sim.shape)
```
